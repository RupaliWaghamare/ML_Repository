### Step 3: Seperate X and Y(SalePrice)

## Step 4: Perform Feature selection
For feature selection pipeline use OrdinalEncoding for Categorical features

```python
# Seperate cat and con featurs
cat = list(X.columns[X.dtypes=='object'])
con = list(X.columns[X.dtypes!='object'])
```

```python
con
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OrdinalEncoder
from sklearn.compose import ColumnTransformer
```

```python
num_pipe1 = Pipeline(steps=[('impute', SimpleImputer(strategy='mean')),
                            ('scaler', StandardScaler())])
```

```python
cat_pipe1 = Pipeline(steps=[('impute', SimpleImputer(strategy='constant', 
                                                     fill_value='Not_Avail')),
                            ('ordinal', OrdinalEncoder())])
```

```python
pre1 = ColumnTransformer([('num', num_pipe1, con),
                          ('cat', cat_pipe1, cat)]).set_output(transform='pandas')
```

```python
X_pre1 = pre1.fit_transform(X) 
X_pre1.head()
```

```python
# Apply feature selection process
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import SequentialFeatureSelector
lr = LinearRegression()
sel = SequentialFeatureSelector(lr, 
                                n_features_to_select='auto', 
                                direction='backward')
sel.fit(X_pre1, Y)
sel_cols = sel.get_feature_names_out()
sel_cols
```

```python
len(sel_cols)
```

```python
sel_cols[0]
```

```python
sel_cols[0].split('__')
```

```python
sel_cols[0].split('__')[1]
```

```python
imp_cols = []
for i in sel_cols:
    s = i.split('__')[1]
    imp_cols.append(s)
```

```python
imp_cols
```

```python
X_sel = X[imp_cols]
X_sel.head()
```

Step 5: Create a main pipeline for Final models


 Categorical features - OneHotEncoder

```python
# Seperate cat_sel and con_sel
cat_sel = list(X_sel.columns[X_sel.dtypes=='object'])
con_sel = list(X_sel.columns[X_sel.dtypes!='object'])
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
num_pipe2 = Pipeline(steps=[('impute', SimpleImputer(strategy='mean')),
                            ('scaler', StandardScaler())])
```

```python
cat_pipe2 = Pipeline(steps=[('impute', SimpleImputer(strategy='constant',
                                                     fill_value='Not_Avail')),
                            ('Ohe', OneHotEncoder(handle_unknown='ignore',
                                                  sparse_output=False))])
```

```python
pre2 = ColumnTransformer([('num', num_pipe2, con_sel),
                          ('cat', cat_pipe2, cat_sel)]).set_output(transform='pandas')
```

```python
pre2
```

```python
X_sel_pre = pre2.fit_transform(X_sel)
X_sel_pre.head()
```

Step 6: Perform train test split

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X_sel_pre, Y, 
                                                test_size=0.2,
                                                random_state=21)
```

Step 7: Build model

```python
from sklearn.linear_model import LinearRegression
model1 = LinearRegression()
model1.fit(xtrain, ytrain)
```

```python
model1.score(xtrain, ytrain)
```

```python
model1.score(xtest, ytest)
```

## Ridge Model

```python
import numpy as np 
params_ridge = {'alpha':np.arange(0.1, 100, 0.1)}
```

```python
params_ridge
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
ridge_model = Ridge()
gscv_ridge = GridSearchCV(estimator=ridge_model, param_grid=params_ridge,
                          cv=5, scoring='r2')
gscv_ridge.fit(xtrain, ytrain)
```

```python
gscv_ridge.best_params_
```

```python
gscv_ridge.best_score_
```

```python
best_ridge = gscv_ridge.best_estimator_
best_ridge
```

```python
best_ridge.score(xtrain, ytrain)
```

```python
best_ridge.score(xtest, ytest)
```

# Lasso (L1) model

```python
gscv_lasso.best_score_
```

```python
gscv_lasso.best_params_
```

```python
best_lasso = gscv_lasso.best_estimator_
best_lasso
```

```python
best_lasso.score(xtrain, ytrain)
```

```python
best_lasso.score(xtest, ytest)
```

Step 8 : Evaluate all the models

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, \
                            mean_absolute_percentage_error, r2_score
```

```python
def evaluate_model(model, x, y):
    ypred = model.predict(x)
    mse = mean_squared_error(y, ypred)
    rmse = mse**(1/2)
    mae = mean_absolute_error(y, ypred)
    mape = mean_absolute_percentage_error(y, ypred)
    r2 = r2_score(y, ypred)
    # Print the results
    print(f'Mean Squared Error : {mse:.2f}')
    print(f'Root Mean Squared Error : {rmse:.2f}')
    print(f'Mean Absolute Error : {mae:.2f}')
    print(f'Mean Absolute Percentage Error : {mape:.4f}')
    print(f'R2 Score : {r2:.4f} ')
```

```python
model1
```

```python
best_ridge
```

```python
best_lasso
```

```python
for i in [model1, best_ridge, best_lasso]:
    print(type(i))
    print(f'Training Results')
    evaluate_model(i, xtrain, ytrain)
    print('\nTesting Results')
    evaluate_model(i, xtest, ytest)
    print('\n=================================================\n')
```
